```python
# Berechnung mit Set
import math
import time
# Log Konfiguration
import logging
logger = logging.getLogger('SetLogger')
# Festlegen des Log Levels - DEBUG, CRITICAL, ERROR, INFO, WARNING, ...
logger.setLevel(logging.DEBUG)
# Erstellen eines FileHandlers - der mode 'w' löscht den Log jedesmal - ansonsten 'a'
handler = logging.FileHandler('Set_logfile.log', mode='w')
# Der Formatter erzeugt die passende Formatierung mit den LogRecord Attributen
# siehe auch https://docs.python.org/3/library/logging.html#formatter-objects
formatter = logging.Formatter('%(asctime)s %(name)s - %(message)s')
handler.setFormatter(formatter)
# Die Handler als Liste an den Logger übergeben
logger.handlers = [handler]



#File Load
def file_to_coords (filename):
    start_time = time.time()
    logger.debug(f"Function: file_to_coords {filename}")
    logger.debug(f"---- entry {start_time}")
    try:
        with open(filename, 'r') as in_file:
            in_list = in_file.read().split()
    except FileNotFoundError:                                                   # Bekannter Fehler ohne Fehlerobjekt
        print ("File ist nicht vorhanden.")
    coordinate_list = []
    for i in range(0,len(in_list),2):
        coord = (float(in_list[i]), float(in_list[i+1]))
        coordinate_list.append(coord)
    end_time = time.time()
    logger.debug(f"---- exit {end_time}")
    logger.debug(f"---- processing time: {end_time - start_time}")
    return coordinate_list

# Simple Distance between two points
def dist_point(a,b):
    return math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)

# Distance of a linestring as list of tupels
def dist_ordered_list (li):
    distance = 0
    for i in range(1,len(li)):
        distance += dist_point(li[i-1],li[i])
    return distance

# Intersects two Lists with keeping the order
# Assumes a common border
def common_border_set (list_1, list_2):
    start_time = time.time()
    logger.debug(f"Function: common_border_set")
    logger.debug(f"---- entry {start_time}")
    set_1 = set(list_1)
    set_2 = set(list_2)
    intersect = set_1.intersection(set_2)  
    # There are no ordered Sets in Python, so we have to keep the order of one List
    common_border = []
    for item in list_1:
        if item in intersect:
            common_border.append(item)
    end_time = time.time()
    logger.debug(f"---- exit {end_time}")
    logger.debug(f"---- processing time: {end_time - start_time}")
    return common_border

# Intersects two Lists with keeping the order
# Assumes a common border
def common_border_list (list_1, list_2):   
    start_time = time.time()
    logger.debug(f"Function: common_border_list")
    logger.debug(f"---- entry {start_time}")
    # There are no ordered Sets in Python, so we have to keep the order of one List
    common_border = []
    for item in list_1:
        if item in list_2:
            common_border.append(item)
    end_time = time.time()
    logger.debug(f"---- exit {end_time}")
    logger.debug(f"---- processing time: {end_time - start_time}")
    return common_border
    
    
#Load two Files
bo_coord = file_to_coords ('Bochum_coord_25832.txt')
do_coord = file_to_coords ('Dortmund_coord_25832.txt')
common_set = common_border_set(bo_coord, do_coord)
common_list = common_border_list(bo_coord, do_coord)
print (f"Bochum  : Stützpunkte: {len(bo_coord)}, Länge: {dist_ordered_list(bo_coord)}")
print (f"Dortmund: Stützpunkte: {len(do_coord)}, Länge: {dist_ordered_list(do_coord)}")
print ("Calculated with sets")
print (f"Common  : Stützpunkte: {len(common_set)}, Länge: {dist_ordered_list(common_set)}")
print ("Calculated with list")
print (f"Common  : Stützpunkte: {len(common_list)}, Länge: {dist_ordered_list(common_list)}")
```

## Lösungsansatz 3 
- Anstelle eine umhüllende Funktion aufzurufen wird eine 'Dekorationsfunktion' erzeug, die eine Referenz auf eine neue Funktion zurückliefert.
- Die zurückgelieferte Funktion wird nun genutzt

```python
# Berechnung mit Set
import math
import time
# Log Konfiguration
import logging
logger = logging.getLogger('SetLogger')
logger.setLevel(logging.DEBUG)
handler = logging.FileHandler('Set_logfile_3.log', mode='w')
formatter = logging.Formatter('%(asctime)s %(name)s - %(message)s')
handler.setFormatter(formatter)
logger.handlers = [handler]

#File Load
def file_to_coords (filename):
    try:
        with open(filename, 'r') as in_file:
            in_list = in_file.read().split()
    except FileNotFoundError:                                                   # Bekannter Fehler ohne Fehlerobjekt
        print ("File ist nicht vorhanden.")
    coordinate_list = []
    for i in range(0,len(in_list),2):
        coord = (float(in_list[i]), float(in_list[i+1]))
        coordinate_list.append(coord)
    return coordinate_list

# Simple Distance between two points
def dist_point(a,b):
    return math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)

# Distance of a linestring as list of tupels
def dist_ordered_list (li):
    distance = 0
    for i in range(1,len(li)):
        distance += dist_point(li[i-1],li[i])
    return distance

# Intersects two Lists with keeping the order
# Assumes a common border
def common_border_set (list_1, list_2):
    set_1 = set(list_1)
    set_2 = set(list_2)
    intersect = set_1.intersection(set_2)  
    # There are no ordered Sets in Python, so we have to keep the order of one List
    common_border = []
    for item in list_1:
        if item in intersect:
            common_border.append(item)
    return common_border

# Intersects two Lists with keeping the order
# Assumes a common border
def common_border_list (list_1, list_2):   
    # There are no ordered Sets in Python, so we have to keep the order of one List
    common_border = []
    for item in list_1:
        if item in list_2:
            common_border.append(item)
    return common_border

# eine "Decorator" Funktion
def time_decorator (function):
    # Rückgabe ist eine veränderte Funktion
    def function_wrapper(*params):
        start_time = time.time()
        logger.debug(f"Function: {function.__name__}")
        logger.debug(f"---- entry {start_time}")
        val = function(*params)
        end_time = time.time()
        logger.debug(f"---- exit {end_time}")
        logger.debug(f"---- processing time: {end_time - start_time}")
        return val
    return function_wrapper

    
#Load two Files
bo_coord = file_to_coords ('Bochum_coord_25832.txt')
do_coord = file_to_coords ('Dortmund_coord_25832.txt')
# Zuweisen einer 'dekorierten' Funktion -> Referenz auf Funktion!
common_border_set = time_decorator (common_border_set)
common_border_list = time_decorator (common_border_list)
# Normale Verwendung der 'dekorierten' Funktion
common_set = common_border_set(bo_coord, do_coord)
common_list = common_border_list(bo_coord, do_coord)
print (f"Bochum  : Stützpunkte: {len(bo_coord)}, Länge: {dist_ordered_list(bo_coord)}")
print (f"Dortmund: Stützpunkte: {len(do_coord)}, Länge: {dist_ordered_list(do_coord)}")
print ("Calculated with sets")
print (f"Common  : Stützpunkte: {len(common_set)}, Länge: {dist_ordered_list(common_set)}")
print ("Calculated with list")
print (f"Common  : Stützpunkte: {len(common_list)}, Länge: {dist_ordered_list(common_list)}")
```
